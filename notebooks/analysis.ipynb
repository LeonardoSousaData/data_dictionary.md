```python
import pandas as pd
from sqlalchemy import create_engine

# Define o usuário do banco
usuario = "root"

# Define a senha do banco; no XAMPP muitas vezes fica vazia
senha = ""

# Define o endereço do servidor local
host = "127.0.0.1"

# Define a porta padrão do MySQL/MariaDB
porta = "3306"

# Define o nome exato do banco criado no DDL
banco = "microsoft_water_finance"

# Cria o motor de conexão entre Python e MariaDB
engine = create_engine(f"mysql+pymysql://{usuario}:{senha}@{host}:{porta}/{banco}")

# Abre uma conexão real para testar acesso ao banco
with engine.connect() as conn:
    print("Conexão realizada com sucesso!")

# Lê a tabela de água para um DataFrame
fact_water = pd.read_sql("SELECT * FROM fact_water;", engine)

# Lê a tabela financeira para um DataFrame
fact_financials = pd.read_sql("SELECT * FROM fact_financials;", engine)

# Lê a tabela de eficiência operacional do data center
fact_datacenter_efficiency = pd.read_sql("SELECT * FROM fact_datacenter_efficiency;", engine)

# Lê a dimensão de anos fiscais
dim_fiscal_year = pd.read_sql("SELECT * FROM dim_fiscal_year;", engine)

# Mostra as primeiras linhas da tabela de água
print(fact_water.head())

# Mostra as primeiras linhas da tabela financeira
print(fact_financials.head())

# Mostra as primeiras linhas da tabela operacional
print(fact_datacenter_efficiency.head())

# Mostra as primeiras linhas da dimensão de anos
print(dim_fiscal_year.head())

# Lê a view analítica completa com KPIs e WUE
df_kpis = pd.read_sql("SELECT * FROM vw_water_finance_dc_kpis;", engine)

# Mostra as primeiras linhas da view
print(df_kpis.head())
```

```python
fact_water_sel = fact_water[[
    "fiscal_year",
    "water_withdrawn_m3",
    "water_consumed_m3",
    "method_flag"
]]

fact_financials_sel = fact_financials[[
    "fiscal_year",
    "revenue_usd"
]]

fact_dc_sel = fact_datacenter_efficiency[[
    "fiscal_year",
    "wue_l_per_kwh"
]]

df_base = fact_water_sel.merge(fact_financials_sel, on="fiscal_year", how="inner")
df_base = df_base.merge(fact_dc_sel, on="fiscal_year", how="left")

print(df_base.to_string(index=False))
```

```python
df_base = df_base.sort_values("fiscal_year").copy()

df_base["withdrawn_intensity"] = df_base["water_withdrawn_m3"] / df_base["revenue_usd"]
df_base["consumed_intensity"] = df_base["water_consumed_m3"] / df_base["revenue_usd"]

df_base["withdrawn_m3_per_usd_billion"] = (
    df_base["water_withdrawn_m3"] / df_base["revenue_usd"]
) * 1_000_000_000

df_base["consumed_m3_per_usd_billion"] = (
    df_base["water_consumed_m3"] / df_base["revenue_usd"]
) * 1_000_000_000

df_base["cycle_consumption_rate"] = (
    df_base["water_consumed_m3"] / df_base["water_withdrawn_m3"]
)

df_base["yoy_revenue_growth"] = df_base["revenue_usd"].pct_change()
df_base["yoy_withdrawn_growth"] = df_base["water_withdrawn_m3"].pct_change()
df_base["yoy_consumed_growth"] = df_base["water_consumed_m3"].pct_change()

print(df_base.to_string(index=False))
```

```python
comparison_cols = [
    "fiscal_year",
    "withdrawn_m3_per_usd_billion",
    "consumed_m3_per_usd_billion",
    "cycle_consumption_rate",
    "yoy_revenue_growth",
    "yoy_withdrawn_growth",
    "yoy_consumed_growth"
]

print("=== KPIs CALCULADOS NO PANDAS ===")
print(df_base[comparison_cols].to_string(index=False))

print("\n=== KPIs DA VIEW SQL ===")
print(df_kpis[comparison_cols].to_string(index=False))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

plt.plot(
    df_base["fiscal_year"],
    df_base["water_withdrawn_m3"],
    marker="o",
    label="Water Withdrawn (m³)"
)

plt.plot(
    df_base["fiscal_year"],
    df_base["water_consumed_m3"],
    marker="o",
    label="Water Consumed (m³)"
)

plt.title("Microsoft Water Withdrawn and Consumed by Fiscal Year")
plt.xlabel("Fiscal Year")
plt.ylabel("Water Volume (m³)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(
    df_base["fiscal_year"],
    df_base["withdrawn_m3_per_usd_billion"],
    marker="o",
    label="Withdrawn Intensity (m³ per US$ 1B)"
)

plt.plot(
    df_base["fiscal_year"],
    df_base["consumed_m3_per_usd_billion"],
    marker="o",
    label="Consumed Intensity (m³ per US$ 1B)"
)

plt.title("Microsoft Water Intensity per US$ 1 Billion Revenue")
plt.xlabel("Fiscal Year")
plt.ylabel("m³ per US$ 1 Billion")
plt.legend()
plt.grid(True)
plt.show()  
```

```python
plt.figure(figsize=(10, 6))

plt.plot(
    df_base["fiscal_year"],
    df_base["yoy_revenue_growth"],
    marker="o",
    label="Revenue YoY Growth"
)

plt.plot(
    df_base["fiscal_year"],
    df_base["yoy_withdrawn_growth"],
    marker="o",
    label="Withdrawn Water YoY Growth"
)

plt.plot(
    df_base["fiscal_year"],
    df_base["yoy_consumed_growth"],
    marker="o",
    label="Consumed Water YoY Growth"
)

plt.title("Microsoft YoY Growth: Revenue vs Water")
plt.xlabel("Fiscal Year")
plt.ylabel("YoY Growth Rate")
plt.legend()
plt.grid(True)
plt.show()
```
